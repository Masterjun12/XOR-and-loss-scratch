```python
import numpy as np
import tensorflow as tf
```

```python
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]], dtype=np.float32)

y = np.array([[0], [1], [1], [0]], dtype=np.float32)
```

```python
w1 = tf.Variable(np.array([[5, -7],
                           [5, -10]], dtype=np.float32))

w2 = tf.Variable(np.array([[3, -3],
                           [4, -21]], dtype=np.float32))

w3 = tf.Variable(np.array([[1], [-3]], dtype=np.float32))

b1 = tf.Variable(np.array([-8, 3], dtype=np.float32))
b2 = tf.Variable(np.array([-2, 4], dtype=np.float32))
b3 = tf.Variable(np.array([-1], dtype=np.float32))
```

```python
linear_output1 = tf.matmul(X, w1) + b1
```

```python
linear_output1
```

```python
activation1 = tf.sigmoid(linear_output1)
activation1
```

```python
linear_output2 = tf.matmul(activation1, w2) + b2
linear_output2
```

```python
activation2 = tf.sigmoid(linear_output2)
activation2
```

```python
linear_output3 = tf.matmul(activation2, w3) + b3
linear_output3
```

```python
y_pred = tf.sigmoid(linear_output3)
```

```python
loss_fn = tf.keras.losses.BinaryCrossentropy()
optimizer = tf.keras.optimizers.Adam()
```

```python
# 훈련 과정
epochs = 1000
for epoch in range(epochs):
    with tf.GradientTape() as tape:
        # 순전파
        linear_output1 = tf.matmul(X, w1) + b1
        activation1 = tf.sigmoid(linear_output1)
        linear_output2 = tf.matmul(activation1, w2) + b2
        activation2 = tf.sigmoid(linear_output2)
        linear_output3 = tf.matmul(activation2, w3) + b3
        y_pred = tf.sigmoid(linear_output3)
        
        # 손실 계산
        loss = loss_fn(y, y_pred)
        
    # 그래디언트 계산 및 역전파
    gradients = tape.gradient(loss, [w1, w2, w3, b1, b2, b3])
    print("\n gradients : ",gradients)
    optimizer.apply_gradients(zip(gradients, [w1, w2, w3, b1, b2, b3]))
    
    # 로그 출력
    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, Loss: {loss.numpy()}")
```

```python
# 최종 예측값 출력
print("\n최종 예측값:\n", y_pred.numpy())
```
